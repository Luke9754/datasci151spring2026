(a) **(Q1)** Compute a geometric formula:

<font size = "3">

The **frustum** of a cone is the shape formed when the top of a cone is cut off parallel to the base, leaving two circular faces of different radii. (Think of a lampshade or a bucket)

The formula for the volume of a frustum of a cone is:


$$V = \frac{1}{3}\pi h (r_1^2 + r_1 r_2 + r_2^2)$$


Where:

- $V$ is the volume of the cylinder
- $r_1$ is the radius of the bottom base
- $r_2$ is the radius of the top base
- $h$ is the height of the frustrum

Calculate $V$ when:

- $r_1 = 6$
- $r_2 = 3$
- $h = 10$


To get full points:
- Define Python variables corresponding to the values of $r_1$, $r_2$, and $h$ above.

- Use your variables to calculate the volume ($V$) using the formula above. This value should also be assigned to a variable.

- Print out the value of the computed volume to the screen.

```python
# Write your answer here
r1 = 6
r2 = 3
h = 10
V = (1/3) * math.pi * h * (r1**2 + r1*r2 + r2**2)
print(V)
```

**(Q6)** Generating random values on a computer.

<font size = "3">

Random* values are often generated on a computer by first generating a uniform random variable between 0 and 1, and then transforming it to another value in some manner. 

For example, to simulate a coin toss we can generate a random number $0 < x < 1$ and check if it is less than $0.5$. If it is, we have flipped "Heads". Otherwise we have flipped "Tails".

This models a "fair" coin, where the probability of flipping Heads is 0.5 (a 50% chance). But we could simulate an "unfair" coin, where the probability of flipping Heads is smaller or larger than 0.5.

Write code that does the following:

- Define a variable `p` that is greater than 0 but less than 1. This is the probability of obtaining Heads. It's your choice what value to set it to.

- Use the function `numpy.random.uniform` to randomly generate a variable `x` that lies between 0 and 1.

- Write code that checks if `x` is smaller than `p`. If so, the code should print the string "The flip was Heads". Otherwise, the code should print the string "The flips was Tails".


<font size = "2">

\* Strictly speaking, the majority of computers can only generate *pseudo-random* numbers, not actual random numbers.

```python
# Your code here 
import numpy as np
p = 0.7
x = np.random.uniform(0, 1)
if x < p:
    print("The flip was Heads")
else:
    print("The flip was Tails")
```
